```python
import pandas as pd
import json
import pprint
```

```python
missouri_df = pd.read_excel("../../app/src/data/CEPR_MO_IL.xlsx", sheet_name="Missouri", dtype="object")
illinois_df = pd.read_excel("../../app/src/data/CEPR_MO_IL.xlsx", sheet_name="Illinois", dtype="object")
# json for STL data must be / has been manually created.
stl_df = pd.read_json("../../app/src/data/stl.json", dtype="object")
```

```python
missouri_df = missouri_df[["Unnamed: 0"
        , "All Workers"
        , "All Frontline Industries"
       ,"Grocery, Convenience, & Drug Stores"
       , "Public Transit"
       , "Trucking, Warehouse, & Postal Service"
       , "Building Cleaning Services"
       , "Health Care"
       , "Childcare & Social Services"]]

illinois_df = illinois_df[["Unnamed: 0"
        , "All Workers"
        , "All Frontline Industries"
       ,"Grocery, Convenience, & Drug Stores"
       , "Public Transit"
       , "Trucking, Warehouse, & Postal Service"
       , "Building Cleaning Services"
       , "Health Care"
       , "Childcare & Social Services"]]
```

```python
missouri_df.rename(columns={"Unnamed: 0": "index"}, inplace=True)
illinois_df.rename(columns={"Unnamed: 0": "index"}, inplace=True)
```

```python
missouri_df.fillna("", inplace=True)
illinois_df.fillna("", inplace=True)
```

```python
missouri_result = missouri_df.to_json(orient="records")
illinois_result = illinois_df.to_json(orient="records")

missouri_parsed = json.loads(missouri_result)
illinois_parsed = json.loads(illinois_result)

missouri_json = json.dumps(missouri_parsed, indent=4)
illinois_json = json.dumps(illinois_parsed, indent=4)
```

```python
missouri_df.to_json(r'../../app/src/data/missouri.json', orient="records")
illinois_df.to_json(r'../../app/src/data/illinois.json', orient="records")
```

```python
with pd.ExcelWriter('../../app/src/data/Essential-Workers.xlsx') as writer:  
    missouri_df.to_excel(writer, sheet_name='Missouri')
    illinois_df.to_excel(writer, sheet_name='Illinois')
    stl_df.to_excel(writer, sheet_name='stl')
```
